```python
# Initialize the driver instance with the options parameter
options = webdriver.ChromeOptions()
options.add_argument("start-maximized")
options.add_argument("--disable-blink-features=AutomationControlled")
options.add_experimental_option("excludeSwitches", ["enable-automation"])
options.add_argument('disable-infobars')
options.add_argument('--disable-extensions')
options.add_argument("--incognito")
prefs = {"profile.default_content_settings.popups": 0}
options.add_experimental_option("prefs", prefs)
#options.add_argument("--headless=new")
driver = webdriver.Chrome(options=options)

# Navigate to the login page
driver.get(product_url)
wait = WebDriverWait(driver, 30)
# Find and fill the username and password fields
username_field = driver.find_element(By.NAME, 'loginKey')
password_field = driver.find_element(By.NAME, 'password')
username_field.send_keys(username1)
password_field.send_keys(password1)

# Find and click the login button
login_button = wait.until(EC.element_to_be_clickable((By.XPATH, '//*[@id="main"]/div/div[2]/div/div/div/div[2]/form/div/div[2]/button')))
login_button = wait.until(EC.element_to_be_clickable((By.CLASS_NAME,'wyhvVD._1EApiB.hq6WM5.L-VL8Q.cepDQ1._7w24N1')))
login_button.click()

# Let us create an empty lists to store the data
data_dict = {
    'Product_Names':[],
    'Product_URLs':[],
    'Current_Price':[],
    'Original_Price':[],
    'Units_Sold':[],
    'Sold_Out':[],
    'Discount_Percentage':[],
}


# Loop through search results
# Lets us define a function to retireve the information we need
def get_page_data(driver, data_dict):
    #suppose we want to selenium to wait for it to exist on the page.
    #Explict waits
    try:
        element = WebDriverWait(driver,30).until(EC.presence_of_element_located((By.CLASS_NAME, 'shop-search-result-view__item.col-xs-2-4')))
        search_results = driver.find_elements(By.CLASS_NAME, 'shop-search-result-view__item.col-xs-2-4')
        
        for result in search_results:
            product_name = result.find_element(By.CLASS_NAME, 'Od5Z3o._2sze40.XwV4kb').text
            try:
                product_url = result.find_element(By.TAG_NAME,'a').get_attribute("href")
                time.sleep(1)
            except NoSuchElementException:
                product_url = 'N/A'
            try:
                current_price = result.find_element(By.CLASS_NAME, "_4m2-Os.e5pdAI").text
            except NoSuchElementException:
                    current_price = "N/A"
            try:
                original_price = result.find_element(By.CLASS_NAME, '_4m2-Os.HtOmnS.giYDgR').text
            except NoSuchElementException:
                original_price = "N/A"
            try:
                units_sold = result.find_element(By.CLASS_NAME,'_2VNMCr._8zpprh').text
            except NoSuchElementException:
                units_sold = "N/A"
            try:
                sold_out = result.find_element(By.CLASS_NAME,'UDFONA').text
            except NoSuchElementException:
                sold_out = "Available"
            try:
                percent_off = result.find_element(By.CSS_SELECTOR,'span.Nv7uAa').text
            except NoSuchElementException:
                percent_off = 'N/A'
                
            # Append the empty lists with the data we have
            data_dict['Product_Names'].append(product_name)
            data_dict['Product_URLs'].append(product_url)
            data_dict['Current_Price'].append(current_price)
            data_dict['Original_Price'].append(original_price)
            data_dict['Units_Sold'].append(units_sold)
            data_dict['Sold_Out'].append(sold_out)
            data_dict['Discount_Percentage'].append(percent_off)

    except NoSuchElementException:
        print("Some elements not found for this product.")
    except Exception as e:
        print(f"An error occurred: {e}")

# Now let us create a loop that will go through all the pages
# Get the initial page content
current_page = 1
while True:
    # Get data from the current page
    print(f"Processing page {current_page}...")
    try:
        WebDriverWait(driver, 30).until(EC.presence_of_all_elements_located((By.CLASS_NAME, 'shop-search-result-view__item.col-xs-2-4')))
        get_page_data(driver,data_dict)
    except StaleElementReferenceException:
        WebDriverWait(driver, 30).until(EC.presence_of_all_elements_located((By.CLASS_NAME, 'shop-search-result-view__item.col-xs-2-4')))
        print("StaleElementReferenceException occurred. Waiting for the page to load again...")
        get_page_data(driver,data_dict)

    # Check if there is a next page link
    next_page_link = driver.find_elements(By.CLASS_NAME, 'shopee-button-outline.shopee-mini-page-controller__next-btn')
    if len(next_page_link) == 0 or 'shopee-button-outline--disabled' in next_page_link[0].get_attribute('class'):
        print("No more pages available.")
        break
    else:
        # Click on the next page link to navigate to the next page
        next_page_link[0].click()
        # Wait for the new page to load
        WebDriverWait(driver, 60).until(EC.presence_of_element_located((By.CLASS_NAME, 'shop-search-result-view__item.col-xs-2-4')))
        time.sleep(10)
    current_page += 1
        
# Create the DataFrame from the collected data
pd.set_option('display.max_colwidth', None)
df = pd.DataFrame(data_dict)

driver.quit()
```

```python
# Now let use manipulate our data
# Let us start by creating a function to extract product ID from the URL
def extract_product_id(url):
    match = re.search(r'\.(\d+)\?', url)
    if match:
        return match.group(1)
    return None

# Apply the function to create the 'Product_ID' column
df['Product_ID'] = df['Product_URLs'].apply(extract_product_id)

df.reset_index(drop=True)

df['Current_Price'] = df['Current_Price'].str.replace('₱','').str.replace(',','').astype(float)
df['Original_Price'] = df['Original_Price'].str.replace('₱','').str.replace('N/A','0')\
.str.replace(',','').astype(float)
df['Discount_Percentage'] = df['Discount_Percentage'].str.replace('N/A','0').str.replace('%','').astype(float).abs()

# Let us add the current date column to our data
current_date = datetime.datetime.now().date()
df['Date'] = current_date
df = df[['Date'] + [col for col in df.columns if col != 'Date']]
# Change the type for product ID
df['Product_ID'] = df['Product_ID'].astype(np.int64)
df['Units_Sold'] = df['Units_Sold'].str.replace('.','').str.replace('N/A','0').str.replace('K','000').str.replace('+','').str.replace('[^\d.]','',regex=True)\
.astype(float)
current_date = datetime.datetime.now().strftime("%Y-%m-%d %H:%M")
df['Date'] = current_date
df = df[['Date'] + [col for col in df.columns if col != 'Date']]
# Change the type for product ID
df['Product_ID'] = df['Product_ID'].astype(np.int64)
# Adding the Brand, country column in as well as the seller center
df['Brand'] = 'huggies'
df['Country'] = login_url.split('/')[2].split('.')[-1].upper()
df['seller_centre'] = 'Shopee'
```
