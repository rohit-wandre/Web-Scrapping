```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support.wait import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import NoSuchElementException, StaleElementReferenceException
import time
import pandas as pd
import re
from datetime import datetime
```

```python
# Initialize the driver instance with the options parameter
options = webdriver.ChromeOptions()
options.add_argument('start-maximized')
options.add_argument('disable-infobars')
options.add_argument('--disable-extensions')
options.add_argument('--no-sandbox')
driver = webdriver.Chrome(options=options)
```

```python
# Shopee login URL
login_url = 'https://shopee.sg/buyer/login?next=https%3A%2F%2Fshopee.sg%2Fhuggies.sg%23product_list'


# Replace with your Shopee username and password
username = 'scrapper2023'
password = 'Scrapper@2023'

# Navigate to the login page
driver.get(login_url)
wait = WebDriverWait(driver, 30)
# Find and fill the username and password fields
username_field = driver.find_element(By.NAME, 'loginKey')
password_field = driver.find_element(By.NAME, 'password')
username_field.send_keys(username)
password_field.send_keys(password)
```

```python
# Find and click the login button
login_button = wait.until(EC.element_to_be_clickable((By.XPATH, '//*[@id="main"]/div/div[2]/div/div/div/div[2]/form/div/div[2]/button')))
login_button.click()

# Let us create an empty lists to store the data
data_dict = {
    'Product_Names':[],
    'Product_URLs':[],
    'Current_Price':[],
    'Original_Price':[],
    'Units_Sold':[],
    'Sold_Out':[],
}

# Loop through search results
# Lets us define a function to retireve the information we need
def get_page_data(driver, data_dict):
    #suppose we want to selenium to wait for it to exist on the page.
    #Explict waits
    try:
        element = WebDriverWait(driver,30).until(EC.presence_of_element_located((By.CLASS_NAME, 'wFlKK5')))
        search_results = driver.find_elements(By.CLASS_NAME, 'wFlKK5')
        
        for result in search_results:
            product_name = result.find_element(By.CLASS_NAME, 'h0HBrE.ckHqor._5Kkvx1').text
            try:
                WebDriverWait(driver,30).until(EC.presence_of_element_located((By.XPATH, '//a[@data-sqe="link"]')))\
                .get_attribute("href")
                time.sleep(1)
            except NoSuchElementException:
                product_url = 'N/A'
            try:
                current_price = result.find_element(By.CLASS_NAME, "_0ZJOIv").text
            except NoSuchElementException:
                    current_price = "N/A"
            try:
                original_price = result.find_element(By.CLASS_NAME, 'zSpiUB.KxcK54._8ENrMD').text
            except NoSuchElementException:
                original_price = "N/A"
            try:
                units_sold = result.find_element(By.CLASS_NAME,'sPnnFI.AGQcTN').text
            except NoSuchElementException:
                units_sold = "N/A"
            try:
                sold_out = result.find_element(By.CLASS_NAME,'UDFONA').text
            except NoSuchElementException:
                sold_out = "Available"
                
            # Append the empty lists with the data we have
            data_dict['Product_Names'].append(product_name)
            data_dict['Product_URLs'].append(product_url)
            data_dict['Current_Price'].append(current_price)
            data_dict['Original_Price'].append(original_price)
            data_dict['Units_Sold'].append(units_sold)
            data_dict['Sold_Out'].append(sold_out)

    except NoSuchElementException:
        print("Some elements not found for this product.")
    except Exception as e:
        print(f"An error occurred: {e}")


# Now let us create a loop that will go through all the pages
# Get the initial page content
initial_page_content = driver.page_source
current_page = 1
while True:
    # Get data from the current page
    print(f"Processing page {current_page}...")
    try:
        get_page_data(driver, data_dict)
    except StaleElementReferenceException:
        print("StaleElementReferenceException occurred. Waiting for the page to load again...")
        WebDriverWait(driver, 30).until(EC.presence_of_element_located((By.CLASS_NAME, 'wFlKK5')))
        get_page_data(driver, data_dict)

    # Check if there is a next page link
    next_page_link = driver.find_elements(By.CLASS_NAME, 'shopee-icon-button--right')
    if len(next_page_link) == 0:
        print("No more pages available.")
        break
    else:
        # Click on the next page link to navigate to the next page
        next_page_link[0].click()
        # Wait for the new page to load
        WebDriverWait(driver, 60).until(EC.presence_of_element_located((By.CLASS_NAME, 'wFlKK5')))
        time.sleep(10)
    
    # Check if the content of the page has changed
    new_page_content = driver.page_source
    if new_page_content == initial_page_content:
        print("No more pages available.")
        break
    
    initial_page_content = new_page_content
    current_page += 1      
        
# Create the DataFrame from the collected data
pd.set_option('display.max_colwidth', None)
df = pd.DataFrame(data_dict)
```
